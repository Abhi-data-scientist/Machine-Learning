```python
# Ordinal Encoding ---> 
# Label Encoding --> column[sub1, sub2, sub3....subn] --> encode 
# Ordinal Encoding --> column[sub1, sub2, sub3....subn] --> Re-Arrange--> encode
```

```python
# LabelEncoder --> When we have singal column
    # --> we can not set sub-category form 
    # has_covid['yes', 'no'] --> labelEncoder --> [1,0]

# OrdinalEncoder --> When we have multiple columns 
                # --> We can set sub-category forms
                # has_covid['yes', 'no'] --> OrdinalEncoder --> [0.1]
```

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('covid_toy.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(columns=['city', 'age', 'fever'], inplace=True)
```

```python
df.head(3)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Female', 'Male'], ['Mild', 'Strong'], ['No', 'Yes']])
```

```python
oe_ar = oe.fit_transform(df)
```

```python
new_df = pd.DataFrame(oe_ar, columns=df.columns)
```

```python
new_df
```
